```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import os
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
def plot_signal(df, num_plots=4):
    plt.figure(figsize=(18, 10), dpi= 60)
    sp_idx = 1
    for i in range(num_plots//2 * 2):
        plt.subplot(num_plots//2, 2, sp_idx)
        plt.plot(df.iloc[:,i])
        plt.title(df.columns[i], fontsize=12)
        sp_idx += 1
    plt.tight_layout() # adjust spacing between subplots to minimize the overlaps
```

## Load Wave data

```python
path_wave = os.path.join('../interview/dataswati/test_python_dataswati_2019/wave_data/')
```

```python
def read_wave_txt(path):
    data = []
    with open(path) as f: 
        for line in f: 
            data.append(float(line))
    return data

list_df = []
for name_file in os.listdir(path_wave):
    if int(name_file.split('.')[0].split('_')[1]) < 76:
        data = read_wave_txt(os.path.join(path_wave, name_file))
    
        list_df.append(pd.DataFrame(data, columns=[name_file.split('.')[0]]))
    
df_wave = pd.concat(list_df, axis=1)
```

```python
df_wave.head()
```

```python
plot_signal(df_wave)
```

## Normalize

```python
scaler = MinMaxScaler()
scaler.fit(df_wave)
df_wave_nor = scaler.transform(df_wave)
df_wave_nor = pd.DataFrame(df_wave_nor, columns = df_wave.columns)
```

```python
df_wave_nor.head()
```

```python
plot_signal(df_wave_nor)
```
